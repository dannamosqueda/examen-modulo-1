```python
#OUTLIERS SIN CUANTITAVIAS
#variables objeto no se aplica, por eso hay que 
#filtro cuantitativo, data como se llama el filtro, al final se convierten en nulos
#datos limpios uniendo cuanti y cuali, siempre axis=1
```

```python
#realizamos diagrama de caja o bigote, esos valores son errores o sufren modificaciones
fig= plt.figure(figsize=(25,18))
data3.plot(kind='box',vert=False)
plt.title('valores atipicos dataframe')
plt.show()
```

```python
y=data3['number_of_cases_boxes_delivered']
p25=y.quantile(0.25) #Q1
p75= y.quantile(0.75) #Q2
iqr =p75 - p25

lsiqr= p75 + 1.5*iqr
liiqr= p25 - 1.5*iqr
print('limite superior permitido',lsiqr)
print('limite inferior permitido', liiqr)
```

```python
outliers1_iqr= data3[(y>lsiqr)|(y<liiqr)]
outliers1_iqr
```

```python
outliers1_iqr=data3[(y<lsiqr)&(y>liiqr)]
outliers1_iqr
```

```python
#realizamos diagrama de caja o bigote, esos valores son errores o sufren modificaciones
fig= plt.figure(figsize=(5,3))
plt.boxplot(outliers1_iqr['number_of_cases_boxes_delivered'])
plt.title('valores atipicos del numero de cajas entregadas')
plt.show() 
```
