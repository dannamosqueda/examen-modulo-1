```python
#carga desde un archivo .csv sin indice
data= pd.read_excel('consumer_mit_lift_lab.xlsx')
data.head(5)
```

```python
valores_nulos=data.isnull().sum()  
valores_nulos
```

```python
data1=data.fillna(method='bfill') #eliminamos  los valores nulos numéricos  
data1
```

```python
valores_nulos=data.isnull().sum()  
valores_nulos
```

```python
data2=data1.copy()
```

```python
#metodo de la mediana variables cuantitativas
data2['300_did_not_find']=data2['300_did_not_find'].fillna(method='ffill')
```

```python
data2['284_additional_products_store']=data2['284_additional_products_store'].fillna(method='ffill')
```

```python
data2['304_additional_services_store']=data2['304_additional_services_store'].fillna(method='ffill')
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#diagrama de caja de bigote de todo el data
fig=plt.figure(figsize=(20,20))
data1.plot(kind='box',vert=False)
plt.title('Valores Atipicos')
plt.show()
```

```python
y=data2['241_age_of_the_interviewee']
p25=y.quantile(0.25)
p75=y.quantile(0.75)
iqr=p75-p25

lsiqr=p75+1.5*iqr
liiqr=p25-1.5*iqr

print('Limite superior permitido', lsiqr)
print('Limite inferior permitido', liiqr)

```

```python
outliers_iqr1=data2[(y>lsiqr)|(y<liiqr)]
outliers_iqr1
```

```python
outliers1_iqr2=data2[(y<lsiqr)&(y>liiqr)] 
outliers1_iqr2
```

```python

fig=plt.figure(figsize=(5,3))
plt.boxplot(outliers1_iqr['241_age_of_the_interviewee'])
plt.title('Valores atipicos de la edad de los entrevistados')
plt.show()
```
